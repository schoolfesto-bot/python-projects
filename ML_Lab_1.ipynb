# Lab 1: Summation Logic
This notebook calculates basic arithmetic as a test for my **ML Environment**.

```python
import math

# This is a classic Machine Learning formula
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Let's test it with a number
test_value = 0
result = sigmoid(test_value)

print(f"When input is {test_value}, the Sigmoid result is: {result}")
```
